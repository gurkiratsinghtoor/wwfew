1.Import library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
data=pd.read_csv("/content/Bengaluru_House_Data.csv")
```

```python
data
```

removing unwanted columns

```python
data.drop(columns=['area_type','availability','society'],inplace=True)
```

```python
data
```

removing **zero , clean tha data , null value , missing value drop

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.location.value_counts()
```

removing leading space

```python
data['location']= data['location'].apply(lambda x:x.strip())
```

```python
data.location.value_counts()
```

data blancing

```python
locations_starts = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
locations_less_then_10_enteries=locations_starts[locations_starts<=10]
```

```python
locations_less_then_10_enteries
```

```python
#replacing less than 10
data['location']= data['location'].apply(lambda x:'others'if x in locations_less_then_10_enteries else x )
```

```python
data['location'].value_counts()
```

```python
data
#location cleanup done
```

```python
data['size'].unique()
```

```python
#taking only numeric value so that we can not depend size columns and new colums bedroom is made
data['bedrooms']=data['size'].apply(lambda x:int(x.split(' ')[0]))
```

```python
data.head()
```

```python
#what are the unique value
data.total_sqft.unique()
```

```python
def clean(sqft):
  tokens = sqft.split('-')
  if len(tokens)==2:
    return (float(tokens[0]) + float(tokens[1])) / 2
  else:
    try:
      return float(sqft)
    except ValueError:
      return None
```

```python
data['total_sqft']=data['total_sqft'].apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
#finding how much per sqft per bedrooms
data['sqft_per_bed']=data['total_sqft']/data['bedrooms']
```

```python
data.sqft_per_bed.describe()
```

```python
#removing less than 300
data=data[data['sqft_per_bed']>=300]
```

```python
data
```

```python
data['price_per_sqft']=data['price']*100000/data['total_sqft']
```

```python
data.price_per_sqft.describe()
```

```python
data3 = data[data['price_per_sqft'] >= 2000]
```

```python
data3 = data3.drop(columns= ['size' ,'sqft_per_bed','price_per_sqft'])
```

```python
data3
```

```python
col_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']), remainder ='passthrough')
```

```python
lr = LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model=make_pipeline(col_trans, scaler, lr)
```

```python
data_input=data3.drop(columns =['price'])
data_output=data3['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_input,data_output,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
new_data_point = pd.DataFrame([['Electronic City Phase II', 1056.0, 2.0, 1.0, 2]], columns=['location','total_sqft','bath','balcony','bedrooms'])
```

```python
predicted_price = model.predict(new_data_point)
print(f"The predicted price for the new data point is: {predicted_price[0]:.2f} Lakhs")
```
